## FIFA 21 - Data Cleaning

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
data.info()
```

```python
# Creating a copy
fifa = data.copy()

fifa.sample(2)
```

**Contract column**

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# Function for extracting contract column
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = 0
        end_date = 0
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_date = int(start_date[:4])
        end_date = int(end_date[:4])
        contract_length = end_date - start_date
    return start_date, end_date, contract_length
```

```python
# Apply the function to Contract column and create a new columns
new_cols = ['Contract Start', 'Contract End', 'Contract Length(yrs)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))
```

```python
# For loop that will loop trough all the column names and insert the data that was extracted 
for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(yrs)']]
```

```python
# Contract category
def categorise_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contracted'
    
# Add contract status column and apply function
fifa.insert(fifa.columns.get_loc('Contract Length(yrs)')+1, 'Contract Status', fifa['Contract'].apply(categorise_status))
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(yrs)', 'Contract Status']]
```

**Height**

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
# Functin for converting the height
def convert_height(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

# Apply function to the height column
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={'Height': 'Height(cm)'})
fifa.columns
```

**Weight**

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
# Functin for converting the height
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds/2.205)
    
# Apply function to the weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns={'Weight': 'Weight(kg)'})
fifa.columns
```
