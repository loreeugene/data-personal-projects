## FIFA 21 - Data Cleaning

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Creating a copy
fifa = data.copy()

fifa.sample(2)
```

```python
fifa['Club'] = fifa['Club'].str.strip()
```

**Contract column**

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# Function for extracting contract column
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = 0
        end_date = 0
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_date = int(start_date[:4])
        end_date = int(end_date[:4])
        contract_length = end_date - start_date
    return start_date, end_date, contract_length
```

```python
# Apply the function to Contract column and create a new columns
new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))
```

```python
# For loop that will loop trough all the column names and insert the data that was extracted 
for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']]
```

```python
# Contract category
def categorise_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contracted'
    
# Add contract status column and apply function
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa['Contract'].apply(categorise_status))
```
