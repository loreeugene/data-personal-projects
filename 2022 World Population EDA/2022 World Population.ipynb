```python
# Checking average population by continent
# df.groupby('Continent').mean()

# Checking average population by continent and sorting by 2022 population from highest to lowest 
df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
```

```python
# Checking about Oceania
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
```

```python
df2.plot()
```

This doesnt makes sense. So, we will transpose it to make the columns as rows and rows as columns.

```python
df2.transpose()
```

```python
# Lets just get a new name as not to overwrite others
df3 = df2.transpose()
df3.plot()
```

Better but doesnt seems look okay. we only need the populations

```python
df.columns
```

```python
# notice that we reverted the years as we want to see from  1970 - 2022

df4 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by="2022 Population", ascending=False)
```

```python
df5 = df4.transpose()
```

```python
df5.plot()
```
